```python
import os
import string
import random
from faker import Faker
from datetime import datetime, timedelta
```

```python
def random_by_ratio(word, ratio):
    rint = random.randint(1,sum(ratio))
    for i, word in enumerate(word):
        if sum(ratio[:i])<rint<=sum(ratio[:i+1]):
            return word
```

```python
def faker_dpots_data(data_sttc, data_nums, ar_num, ar_type, seq):
    # ar_path = r'/home/appuser/out_data/simula_data/{data_sttc}/44-BoEing-AR_AR_HIST_{ar_num}-1G-{data_sttc}.txt'
    # faker_path = r'/home/appuser/out_data/simula_data/{data_sttc}/44-BoEing-AR_{ar_type}_SEM_INFO-1G-{data_sttc}.txt'
    ar_path = r'C:\Users\stone0\Desktop\work\智能营销管理系统\30对公存贷款假数据\tmp_data\44-BoEing-AR_AR_HIST_' + ar_num + '-1G-20191213.txt'
    faker_path = r'C:\Users\stone0\Desktop\work\智能营销管理系统\30对公存贷款假数据\\' + ar_num + ar_type + 'test.txt'
    # 预存已有合约号
    ar_file = open(ar_path, mode='r', encoding='utf-8')
    ar_list = []
    for i in range(1, data_nums//3+1):
        num_seq_ar = ar_file.readline()[6:23]
        ar_list.append(num_seq_ar)
    ar_file.close()
    # 开始生成假数据
    openFile = open(faker_path, mode='w+', encoding='utf-8')
    print('开始生成' + ar_type + '数据...')
    for i in range(1, data_nums + 1):
        cod_fn_ent = '0103'
        cod_pro_cod = '44'
        num_seq = '0'
        num_sub_ar = ''
        cod_pd_line = '01'
        date_strt_lcs = data_sttc
        num_chk = '?????'
        cod_msk_ar_sem = '0' * 50
        txt_rem = ''
        # DPOS
        if ar_num == '50':
            num_ar_sem = '44' + str(f.random_number(4)).zfill(4) + random_by_ratio(['010400', '140400'], [600, 20]) + str(f.random_number(5)).zfill(5)
            if i <= data_nums//3:
                num_seq_ar = ar_list[i-1]
            else:
                num_seq_ar = '44' + str(f.random_number(4)).zfill(4) + random_by_ratio(['50', '01'], [600, 50]) + str(f.random_number(1)) + '000' + str(f.random_number(5)).zfill(5)
            id_grp_pd = random_by_ratio(['07', '09'], [650, 1])
            years = range(1999,2020)
            year_ratio = [430, 40, 40, 30, 5, 20, 20, 9, 10, 4, 2, 7, 5, 6, 3, 1, 1, 1, 8, 3, 9]
            cod_lcs_ar_sem = random_by_ratio(['01', '02', '03'], [400, 20, 200])
        else:
            # DPOT
            if i <= data_nums//3:
                num_ar_sem = ar_list[i-1]
            else:
                num_ar_sem = '44' + str(f.random_number(4)).zfill(4) + '51' + str(f.random_digit()) + '0000' + str(f.random_number(4)).zfill(4)
            num_seq_ar = num_ar_sem
            id_grp_pd = random_by_ratio(['31', '30'], [90, 10])
            years = range(2016,2020)
            year_ratio = [90, 10, 3, 1]
            cod_lcs_ar_sem = random_by_ratio(['01', '03'], [40, 60])
        year_list = []
        for y in years:
            year_list.append(str(y))
        year_list.reverse()
        date_reg = random_by_ratio(year_list, year_ratio) + datetime.strftime(f.date_between(start_date="-20y", end_date=datetime.strptime(data_sttc, '%Y%m%d')),'%m%d')
        # 结果数据
        data_line = cod_fn_ent + seq + cod_pro_cod + seq + num_ar_sem + seq + num_seq + seq + num_seq_ar + seq + num_sub_ar + seq + cod_pd_line + seq \
                   + id_grp_pd + seq + date_reg + seq + cod_lcs_ar_sem + seq + date_strt_lcs + seq + num_chk + seq + cod_msk_ar_sem + seq + txt_rem
    #     print(data_line)
        openFile.write(data_line + '\n')
        print(i, end = '...')
    ar_file.close()
    openFile.close()
    print('OK.')
```

```python
def faker_marg_data(data_sttc, data_nums, seq):
    # ar_path = r'/home/appuser/out_data/simula_data/{data_sttc}/44-BoEing-AR_AR_HIST_{ar_num}-1G-{data_sttc}.txt'
    # faker_path = r'/home/appuser/out_data/simula_data/{data_sttc}/44-BoEing-AR_{ar_type}_SEM_INFO-1G-{data_sttc}.txt'
    ar_path = r'C:\Users\stone0\Desktop\work\智能营销管理系统\30对公存贷款假数据\tmp_data\44-BoEing-AR_AR_HIST_62-1G-20191213.txt'
    faker_path = r'C:\Users\stone0\Desktop\work\智能营销管理系统\30对公存贷款假数据\62MARGtest.txt'
    ar_file = open(ar_path, mode='r', encoding='utf-8')
    ar_list = []
    for i in range(1, data_nums//3+1):
        num_seq_ar = ar_file.readline()[6:23]
        ar_list.append(num_seq_ar)
    ar_file.close()

    openFile = open(faker_path, mode='w+', encoding='utf-8')
    print('开始生成数据...')
    for i in range(1, data_nums + 1):
        cod_fn_ent = '0103'
        if i <= data_nums//3:
            num_ar_sem = ar_list[i-1]
        else:
            num_ar_sem = '44' + str(f.random_number(4)).zfill(4) + '62' + str(f.random_digit()) + '0000' + str(f.random_int(29, 3000)).zfill(4)
        num_seq = '0'
        num_seq_ar = num_ar_sem
        num_sub_ar = ''
        date_reg = '2019' + datetime.strftime(f.date_between(start_date="-20y", end_date=datetime.strptime('20191213', '%Y%m%d')),'%m%d')
        cod_lcs_ar_sem = random_by_ratio(['01', '03'], [500, 5])
        date_strt_lcs = '20191213'
        num_chk = '?????'
        cod_msk_ar_sem = '0' * 50
        txt_rem = ''
        data_line = cod_fn_ent + seq + seq + num_ar_sem + seq + num_seq + seq + num_seq_ar + seq + num_sub_ar + seq \
                   + date_reg + seq + cod_lcs_ar_sem + seq + date_strt_lcs + seq + num_chk + seq + cod_msk_ar_sem + seq + txt_rem
    #     print(data_line)
        openFile.write(data_line + '\n')
        print(i, end = '...')
    openFile.close()
    print('OK.')
```
