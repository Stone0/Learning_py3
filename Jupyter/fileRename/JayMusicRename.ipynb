```python
import os
import sys
```

```python
jay_path = 'C:\个人\music\Jay\\'
jay_list = os.listdir(jay_path)
for i in jay_list:
    print(i)
    
    for j in os.listdir(jay_path + i):
        oldname = jay_path + i + '\\' + j
        print(' |-oldname: ' + j)
        newname = jay_path + i + '\\' + j.replace(' ','')
        # 文件名有空格
        if j.find(' ') != -1:
            print('需要删除空格')
            os.rename(oldname, newname)
            print('   |-newname: ' + newname)
        else:
            continue
```
