```python
import codecs
filename = "hadoopforeword.txt"
with codecs.open(filename, encoding='utf8') as f:
    mytext = f.read()
mytext[:500]
```

```python
from wordcloud import WordCloud
# wordcloud = WordCloud().generate(mytext)
```

```python
from PIL import Image
import numpy as np
# read the mask / color image taken from
alice_coloring = np.array(Image.open("wsc.jpg"))
```

```python
from wordcloud import STOPWORDS
# 设置停用词
stopwords = set(STOPWORDS)
# stopwords.add("said")
```

```python
# 通过mask参数设置词云形状
wordcloud = WordCloud(background_color="black", max_words=500, mask=alice_coloring,
               stopwords=stopwords, max_font_size=40, random_state=42).generate(mytext)
```

```python
from wordcloud import ImageColorGenerator
# create coloring from image
image_colors = ImageColorGenerator(alice_coloring)
```

```python
import matplotlib.pyplot as plt
plt.imshow(WordCloud().generate(mytext), interpolation='bilinear')
plt.axis("off")
plt.figure()
```

```python
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
# 我们还可以直接在构造函数中直接给颜色
# 通过这种方式词云将会按照给定的图片颜色布局生成字体颜色策略
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.figure()
plt.imshow(alice_coloring, cmap=plt.cm.gray, interpolation="bilinear")
plt.axis("off")
plt.figure()
plt.show()
```
